### 1 预处理数据

```python
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
print(iris_dataset.keys())
print(iris_dataset['target_names'])
print(iris_dataset['feature_names'])
```

```python
print('first 5 rows:\n{}'.format(iris_dataset['data'][:5]))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state = 0
) # 把数据集随机打乱，分成75%训练和25%测试数据
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### 2 观察数据

```python
# 创建DataFrame

import pandas as pd
import matplotlib.pyplot as plt

iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names)

# 画出图像
grr = pd.plotting.scatter_matrix(iris_dataframe, c = y_train, figsize = (15, 15), marker = 'o', hist_kwds = {'bins': 20}, s = 60, alpha = .8, cmap = plt.get_cmap('jet'))
```

### 3 用kNN算法构建模型

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

### 4 作出预测

```python
import numpy as np

X_new = np.array([[5, 2.9, 1, 0.2]])
prediction = knn.predict(X_new)
print(prediction, iris_dataset['target_names'][prediction])
```

### 5 评估模型

```python
y_pred = knn.predict(X_test)
print(y_pred)

print('test set score: {:.4f}'.format(np.mean(y_pred == y_test)))
```
