```python
# 两种画图接口

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn-whitegrid')
x = np.linspace(-10, 10, 1000)

# 1 matlab风格接口

plt.figure() # 创建图形

# 创建两个子图中的第一个，设置坐标轴
plt.subplot(2, 1, 1) # (行、列、子图编号)
plt.plot(x, np.sin(x))

# 创建两个子图中的第二个，设置坐标轴
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

# 用plt.gcf()（获取当前图形）和plt.gca()（获取当前坐标轴）来查看具体信息。
```

```python
# 2 面向对象接口

# 先创建图形网格
# ax是一个包含两个Axes对象的数组
fig, ax = plt.subplots(2)
# 在每个对象上调用plot()方法
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x - 0), color='blue') # 标准颜色名称
plt.plot(x, np.sin(x - 1), color='g') # 缩写颜色代码（rgbcmyk）
plt.plot(x, np.sin(x - 2), color='0.75') # 范围在0~1的灰度值
plt.plot(x, np.sin(x - 3), color='#FFDD44') # 十六进制（RRGGBB，00~FF）
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB元组，范围在0~1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML颜色名称
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 3, linestyle='dashed')
plt.plot(x, x + 6, linestyle='dashdot')
plt.plot(x, x + 9, linestyle='dotted')

# 你可以用下面的简写形式
# plt.plot(x, x + 12, linestyle='-') # 实线
# plt.plot(x, x + 15, linestyle='--') # 虚线
# plt.plot(x, x + 18, linestyle='-.') # 点划线
# plt.plot(x, x + 21, linestyle=':') # 实点线
```

```python
plt.plot(x, x + 0, '-g') # 绿色实线
plt.plot(x, x + 3, '--c') # 青色虚线
plt.plot(x, x + 6, '-.k') # 黑色点划线
plt.plot(x, x + 9, ':r'); # 红色实点线
```
